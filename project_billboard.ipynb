# Project | Billboard Hot 100

You have find data on the internet about currently popular songs. Billboard maintains a weekly Top 100 of "hot" songs here: https://www.billboard.com/charts/hot-100.

```python
# Loading the library
from bs4 import BeautifulSoup
import requests
import pandas as pd
from random import sample, random
```

```python
#Import library
import sys
from credentials import *
```

```python
import spotipy
import json
from spotipy.oauth2 import SpotifyClientCredentials

#Initialize SpotiPy with user credentias
sp = spotipy.Spotify(auth_manager=SpotifyClientCredentials(client_id= Client_ID, client_secret= Client_Secret))
```

```python
# Getting the URL
url = "https://www.billboard.com/charts/hot-100"

# Getting the html of the webpage
response = requests.get(url)
response.status_code
```

```python
# Parsing the html code
soup = BeautifulSoup(response.content, "html.parser")
```

```python
# Creating empty lists for the new dataframe for the song, artist and ranking
song = []
artist = []
rank =[]
```

```python
# Song column
num_iter = len(soup.find_all("span","chart-element__information__song text--truncate color--primary"))

for i in range(num_iter):
    song.append(soup.find_all("span","chart-element__information__song text--truncate color--primary")[i].get_text())
```

```python
# Artist column
num_iter = len(soup.find_all("span","chart-element__information__artist text--truncate color--secondary"))

for i in range(num_iter):
    artist.append(soup.find_all("span","chart-element__information__artist text--truncate color--secondary")[i].get_text())
```

```python
# Rank column
num_iter = len(soup.find_all("span","chart-element__rank__number"))

for i in range(num_iter):
    rank.append(soup.find_all("span","chart-element__rank__number")[i].get_text())
```

```python
# Creating the top_music dataframe
top_music = pd.DataFrame({"rank":rank,"song":song,"artist":artist})
```

```python
# Having an extra column for lower characters
song_lower = []
for i in top_music['song']:
    song_lower.append(i.lower())
```

```python
# Creating the top_music dataframe
top_music = pd.DataFrame({"rank":rank,"song":song,"song_lower":song_lower,"artist":artist})
top_music
```

```python
#List of songs from Spotify
music_list = pd.read_csv('music_listing.csv')
music_list
```

```python
music_list.columns
```

```python
music_list = music_list.drop(['Unnamed: 0'], axis=1)
```

```python
#Final function: insert music you are searching for, and then a recommendation from that chart
import random

def music_charts():
    music_search = input("Enter the music you are searching: ")
    if music_search in list(top_music['song']):
        print("It's in the Hot Billboard")
        print()
        print("Other three songs you might like: ")
        print(random.sample(list(top_music['song']),3))
    elif music_search in list(top_music['song_lower']):
        print("It's in the Hot Billboard")
        print()
        print("Other three songs you might like: ")
        print(random.sample(list(top_music['song_lower']),3))
    else:
        print("It's not in the Hot Billboard")
        import pickle
        
        with open('model.pkl', 'rb') as file:
            kmeans = pickle.load(file)
        
        with open('scaler.pkl', 'rb') as file:
            scaler = pickle.load(file)
        song = sp.search(q=music_search, limit=1) 
        song = song["tracks"]["items"][0]["uri"] 
        features = sp.audio_features(song)[0]
        df=pd.DataFrame(data = features, index=[0])[['danceability', 'energy', 'key', 'loudness', 'mode', 'speechiness','acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo']]
        #print(features)
        display(df)
        scaled_features = scaler.transform(df)
        cluster_song = kmeans.predict(scaled_features)[0]
        recomended_song = music_list[music_list['cluster'] == cluster_song].sample()
        print(recomended_song)
```

```python
# an example of a song in the dataframe
music_charts()
```
